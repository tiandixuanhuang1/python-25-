## 第0010题 生成类似验证码之类的东西

```python
from PIL import Image, ImageDraw, ImageFont, ImageFilter

import random

def md_char():
    return chr(random.randint(65, 90)) #字母的ASCll码值在65-90之间

def my_color(): #定义随机的颜色（*， *， *）表示三原色的数值
    return (random.randint(64, 255), random.randint(64, 255), random.randint(64, 255))

def my_color2():
    return (random.randint(32, 127), random.randint(32, 127), random.randint(32, 127))

width = 60*4
height = 60

image = Image.new('RGB', (width, height), (255, 255, 255))

font = ImageFont.truetype("Arial.ttf", 36)

draw = ImageDraw.Draw(image) #构造绘画对象

for i in range(width):
    for j in range(height):
        draw.point((i, j), fill = my_color()) #转换图像的像素值
        
for t in range(4):
    draw.text((60*t + 10, 10), md_char(), font = font, fill = my_color2())
#draw.text(address, string, font, color) address = (x,y)，表示x坐标与y坐标
    
image = image.filter(ImageFilter.BLUR)# 这一步。。。。无语
image

# import numpy as np
# print(np.array(image)) #使用np中array的方法可以将原来的Image文件转换成numpy中3维数组的形式
```

```python
from PIL import Image, ImageDraw, ImageFont,ImageFilter

def add_num(img):
    draw = ImageDraw.Draw(img)
    fillcolor = "#ff0000"
    
    myfont = ImageFont.truetype('Arial.ttf',size=100)
    
    width, height = image.size #确定图片的长宽
    draw.text((width -100, 500), '2',font = myfont, fill = fillcolor)
    img.save("result.jpg",'jpeg')
    return 0

image = Image.open('WechatIMG1.jpeg')
print(image.format,image.size,image.mode)
add_num(image)
```

## 第 0011 题： 敏感词文本文件 filtered_words.txt，里面的内容为以下内容，当用户输入敏感词语时，则打印出 Freedom，否则打印出 Human Rights。

```python
import re

def open_txt(file):
    with open(file) as f:
        data = f.read()
        list_data = data.split('\n')
        return list_data


def is_sensitive(sensitive_words):   
    a = 'y'
    while a=='y':
        words = input("please import your words:")
        if words in sensitive_words:
            print("Freedom")
            a = input("Would you want it again? y/n: ")

        else:
            print("Human Rights!")
            a = input("Would you want it again? y/n: ")
    print("Thank for you playing! ")
        

file = 'filtered_words.txt'
sensitive_words = open_txt(file)
is_sensitive(sensitive_words)
```
